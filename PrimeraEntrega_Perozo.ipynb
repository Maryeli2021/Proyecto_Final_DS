# ANALISIS DEL DISEÑO DE FRACTURAS EN POZOS NO CONVENCIONALES

## 📙 **INTRODUCCIÓN**
El proyecto de Ciencia de Datos que a continuación se presenta esta enfocado en datos reales de pozos petroleros realizados en la Republica Argentina en donde se ha aplicado el Fracturamiento Hidráulico como técnica de estimulación para posibilitar o aumentar la extracción de gas y petróleo del subsuelo. El Fracturamineto Hidrulico, también denominado Fracking o Estimulación Hidráulica es el proceso de bombear un fluido dentro del pozo a una tasa de inyección muy alta, con el objetivo de generar uno o varios canales de flujo de elevada permeabilidad de modo que supere la resistencia de la roca y abra una fractura controlada en el fondo del pozo para colocar un elemento de empaque (arena) que permita incrementar la conductividad de la formación y, por ende, el flujo de fluidos hacia el pozo. 
Este método empezó a aplicarse industrialmente en 1949 en los Estados Unidos y se estima que en 2010 esta técnica estaba presente en aproximadamente el 60 % de los pozos de extracción en uso. En Argentina se comenzó a usar esta técnica en la formación Vaca Muerta, ubicada en la provincia de Neuquén, a partir del año 2011. Tambien ha sido usada en Santa Cruz, en cercanías a Los Perales. Argentina ocupa a nivel mundial el segundo puesto en recursos de gas no convencional y el cuarto en petróleo.


### *TEMATICA*
En la realización de un proyecto que implique una inversion es  muy importante estudiar todos los aspectos que nos permitan minimizar los costos del mismo. En un proyecto de fracturamiento especificamente, donde la inversion inicial es altisima y los riesgos son un factor muy relevante, tratar de minimizar los costos y por lo tanto los riesgos implicados en el mismo es un paso importantisimo para la toma de deciones mas acertivas.
Existen diversas variables a tomar en cuenta a la hora de planear un proyecto de fracturamiento, entre las mas importantes a analizar es la relacion permebilidad/porosidad (K/Phi). Esta relacón nos permite identificar el tipo de diseño de fractura más conveniente. En general, si se hace una evaluación de tratamientos hechos a nivel mundial, se puede decir que en función de la permeabilidad y de la porosidad de la formación, hay un rango de estimulaciones que se pueden ejecutar.Sin embargo en este conjunto de datos con los que trabajeremos en este proyecto no tenemos acceso a los mismos ya que dichas variables forman parte del conjunto de datos calificados como información privada de las empresas operadoras de los yacimientos. Otra variable importante en un proeycto de este tipo es el contenido de materia orgánica en la roca, a partir del cual se estima la calidad de petróleo que se puede extraer. Aun cuando el valor númerico de este dato no este disponible, se sabe por estudios geologicos públicados anteriormente que este valor es alto en la Formación Vaca Muerta, lo que la hace atractiva para que empresas nacionales e internacionales deseen invertir en proyectos de extraccion de petróleo no convencional en Argentina. Por lo tanto en este análisis estudiaremos el resultado que han tenido las fracturas medidos desde la producción de los pozos para tratar de determinar particularmnente que variables tienen mayor influencia en el diseño de las fracturas y asi tratar de aplicar los mismos parámetros de diseño en pozos a proponer.

El conjunto de datos con el cual trabajaremos en este proyecto muestra el historico de pozos no convencionales terminados en Argentina. **VARIABLES DEL DISEÑO DE LA FRACTURA**

* id_base_fractura_adjiv: número de identificación de la fractura.
* idpozo:número de identificación del pozo.
* sigla: nombre completo del pozo
* cuenca: cuenca donde se encuentra el pozo.
* areapermisoconcesion: area de concesión donde se encuentra el pozo.
* yacimiento: yacimiento donde se encuentra el pozo.
* formacion_productiva: formación desde la cual se extra el fluido (petróleo/gas) de producción.
* tipo_reservorio: tipo de reservorio (convencional, no convencional).
* subtipo_reservorio: subtipo de reservorio (tight/shale).
* longitud_rama_horizontal_m: longitud del tramo del pozo donde se realiza la fractura, medido en metros.
* cantidad_fracturas: cantidad de fractruras realizadas en el pozo.
* tipo_terminacion: técnica con la que se punza (abrir el agujero en el pozo) para realizar la fractura en el yacimiento.
* arena_bombeada_nacional_tn: cantidad de arena que se bombea hacia el pozo, del tipo nacional, medida en toneladas.
* arena_bombeada_importada_tn:	cantidad de arena que se bombea hacia el pozo, del tipo impotada, medida en toneladas.
* agua_inyectada_m3: cantidad de agua inyectada al pozo para producir la fractura, medida en metros cúbicos.
* co2_inyectado_m3: cantidad de CO2 inyectado al pozo para producir la fractura, medido en metros cúbicos.	
* presion_maxima_psi: presión máxima para producir la fractura medida en psi (libras sobre pulgadas cuadradas).
* potencia_equipos_fractura_hp: potencia de los equipos usados para producir la fractura, medidos en hp (caballos de fuerza).
* fecha_inicio_fractura: fecha donde se inicio la fractura en el pozo, especificando dia de la semana, dia del mes, mes y año.
* fecha_fin_fractura:fecha donde se finalizo la fractura en el pozo, especificando dia de la semana, dia del mes, mes y año.
* fecha_data: fecha a la cual fue medido el dato reportado, expresado en dia y hora especifica.
* anio_if: año en el cual se inicio la fractura.
* mes_if: mes en el cual se inicio la fractura.
* anio_ff: año en el que finalizo la fractura.
* mes_ff: mes en el que finalizo la fractura.
* anio_carga:
* mes_carga:
* empresa_informante: empresa que realizo la fractura.
* mes: mes en el que se realizo la fractura.
* anio: año en el que se realizo la fractura.

**VARIABLES DE PRODUCCIÓN**

* GOR: Relación gas/petróleo producido.
* Pet_1A: cantidad acumulada de petróleo durante el primer año de producción del pozo.
* Gas_1A: cantidad acumulada de gas durante el primer año de producción del pozo.	
* Pet_2A: cantidad acumulada de petróleo despues de dos años de producción del pozo.	
* Gas_2A: cantidad acumulada de gas despues de dos años de producción del pozo.	
* ACUM_Pet: cantidad acumulada de petróleo por pozo.	
* ACUM_Gas: cantidad acumulada de gas por pozo.	
* TipoPozo: tipo de pozo (gasifero, petrolifero, sumidero, otro).	
* TipoExtraccion: método de extracción del fluido (surgencia natural, cavidad progresiva, gas lift, bombeo mecanico).		
* Diseno_Pozo: tipo de diseño del pozo (horizontal, vertical o dirigido)

Una vez que ya tenemos los datos mas depurados, procedemos a crear nuevas variables de uso comun en la industria para facilitar el analisis de la correlación de los distintos parametros de fractura con la variable "Acumulada de Petróleo". A partir de este gráfico podemos observar que debemos agrupar algunas variables que definen a las fracturas para poder establecer una mejor correlación con la Acumulada de Petróleo (ACUM_Pet). Antes vamos a calcular con cuales variables referente a las fracturas se observa alguna correlación y para poder comparar los datos a la misma condicion de tiempo, analicemos usando la Acumulada al Primer año, de esta manera podremos observar bien la correlacion entre las variables basados en un mismo periodo de tiempo.

```python
Corr_PET_1A=DS_OGnum.corr()['Pet_1A'][:-1]
Corr_PET_List_1A=Corr_PET_1A[abs(Corr_PET_1A)> 0.1].sort_values(ascending=False)
print("Hay {} variables con correlación mayor a un 10% con Pet_1A:\n{}".format(len(Corr_PET_List_1A),Corr_PET_List_1A))
```

La variable LRH es la variable de los datos de fractura con la cual encontramos una mejor correlacion con la Acumulada de Petróleo con un 15.9%. Asi que corroboramos lo que habiamos mencionado anteriormente con el clustermap, tendriamos que crear varibles nuevas agrupando las que ya tenemos para asi lograr una mejor correlación con las varibles de produción.

### OPTIMIZACIÓN DE HIPERPARAMETROS

Para realizar esta optimización primero buscamos el número optimo de clusters a usar. Aplicaremos el método de Silhouette.

```python
#Buscamos el numero optimo de clusters

from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score,silhouette_score,silhouette_samples,calinski_harabasz_score
# diccionarios vacios para ser almacenados los resultados
sse,db,slc,cal = {}, {}, {}, {}
for k in range(2, 6):
    # semilla de 10 para asegurar reproducibilidad
    kmeans = KMeans(n_clusters=k, max_iter=1000,random_state=10).fit(X) # Algoritmo K means
    if k == 4: labels = kmeans.labels_
    clusters = kmeans.labels_ # Extraccion de labels
    sse[k] = kmeans.inertia_ # Inercia: Suma de las distancias de las muestras a su centroide mas cercano
    db[k] = davies_bouldin_score(X,clusters) # Calculo del indice de Davies Bouldin
    slc[k] = silhouette_score(X,clusters) # Calculo del indice de Silhouette
    cal[k] = calinski_harabasz_score(X,clusters)  # Calculo del indice de Calinski-Harabasz
```

```python
plt.figure(figsize=(15,10))
plt.plot(list(slc.keys()), list(slc.values()))
plt.xlabel("Número de clusters")
plt.ylabel("Score de Silhouette")
plt.show()
```

De este grafico podemos observar que el número optimo de clusters seria 2, ya que justo en ese punto se observa como la curva desciende.

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    """Generate a simple plot of the test and training learning curve"""
    plt.figure(figsize=(15,8))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

g = plot_learning_curve(knn_grid_search.best_estimator_,"\n K-Neighbors learning curves",X_train,y_train,cv=cv)
g = plot_learning_curve(rf_grid_search.best_estimator_,"\n Random Forest learning curves",X_train,y_train,cv=cv)
```

A partir de estos gráficos podemos concluir que las curvas de entrenamiento del modelo K-nearest neighbour muestran un ajuste óptimo en los primeros 100 nodos (K=100), es decir que tenemos un win to win, podemos ver que se gana presición tanto en el set de entrenamiento como en el de prueba. A partir de este punto, se gana presición en el test pero se pierde en el entrenamiento. Podemos decir entonces que en modelo predice sin fundamento, por lo tanto necesita entrenamiento.

Con respecto al modelo Random Forest las curvas se muestran paralelas a aprtir de los 150 nodos; se observa que las curvas de entrenamiento muestran un acierto alto pero las de prueba muestran poco acierto. En esta última podria interpretarse que el modelo tiende a sobreajustarse, ya que no logra generalizar los datos y la curva de Training Score no nos muestra "sombra".

Por lo tanto elejimos el modelo K-nearest Neighbors para futuras predicciones.
