# ANALISIS DEL DISEÑO DE FRACTURAS EN POZOS NO CONVENCIONALES

## 📙 **INTRODUCCIÓN**
El proyecto de Ciencia de Datos que a continuación se presenta esta enfocado en datos reales de pozos petroleros realizados en la Republica Argentina en donde se ha aplicado el Fracturamiento Hidráulico como técnica de estimulación para posibilitar o aumentar la extracción de gas y petróleo del subsuelo. El Fracturamineto Hidrulico, también denominado Fracking o Estimulación Hidráulica es el proceso de bombear un fluido dentro del pozo a una tasa de inyección muy alta, con el objetivo de generar uno o varios canales de flujo de elevada permeabilidad de modo que supere la resistencia de la roca y abra una fractura controlada en el fondo del pozo para colocar un elemento de empaque (arena) que permita incrementar la conductividad de la formación y, por ende, el flujo de fluidos hacia el pozo. 
Este método empezó a aplicarse industrialmente en 1949 en los Estados Unidos y se estima que en 2010 esta técnica estaba presente en aproximadamente el 60 % de los pozos de extracción en uso. En Argentina se comenzó a usar esta técnica en la formación Vaca Muerta, ubicada en la provincia de Neuquén, a partir del año 2011. Tambien ha sido usada en Santa Cruz, en cercanías a Los Perales. Argentina ocupa a nivel mundial el segundo puesto en recursos de gas no convencional y el cuarto en petróleo.

Existen diversas variables a tomar en cuenta a la hora de planear un proyecto de fracturamiento, entre las mas importantes a analizar es la relacion permebilidad/porosidad (K/Phi). Esta relacón nos permite identificar el tipo de diseño de fractura más conveniente. En general, si se hace una evaluación de tratamientos hechos a nivel mundial, se puede decir que en función de la permeabilidad y de la porosidad de la formación, hay un rango de estimulaciones que se pueden ejecutar.Sin embargo en este conjunto de datos con los que trabajeremos en este proyecto no tenemos acceso a los mismos ya que dichas variables forman parte del conjunto de datos calificados como información privada de las empresas operadoras de los yacimientos. Otra variable importante en un proeycto de este tipo es el contenido de materia orgánica en la roca, a partir del cual se estima la calidad de petróleo que se puede extraer. Aun cuando el valor númerico de este dato no este disponible, se sabe por estudios geologicos públicados anteriormente que este valor es alto en la Formación Vaca Muerta, lo que la hace atractiva para que empresas nacionales e internacionales deseen invertir en proyectos de extraccion de petróleo no convencional en Argentina. * id_base_fractura_adjiv: número de identificación de la fractura.

* cuenca: cuenca donde se encuentra el pozo.

* agua_inyectada_m3: cantidad de agua inyectada al pozo para producir la fractura, medida en metros cúbicos.
* co2_inyectado_m3: cantidad de CO2 inyectado al pozo para producir la fractura, medido en metros cúbicos.

* potencia_equipos_fractura_hp: potencia de los equipos usados para producir la fractura, medidos en hp (caballos de fuerza).
* fecha_inicio_fractura: fecha donde se inicio la fractura en el pozo, especificando dia de la semana, dia del mes, mes y año.
* fecha_fin_fractura:fecha donde se finalizo la fractura en el pozo, especificando dia de la semana, dia del mes, mes y año.
* fecha_data: fecha a la cual fue medido el dato reportado, expresado en dia y hora especifica.
* anio_if: año en el cual se inicio la fractura.
* mes_if: mes en el cual se inicio la fractura.
* anio_ff: año en el que finalizo la fractura.
* mes_ff: mes en el que finalizo la fractura.
* anio_carga:
* mes_carga:
* empresa_informante: empresa que realizo la fractura.
* mes: mes en el que se realizo la fractura.
* anio: año en el que se realizo la fractura.

**VARIABLES DE PRODUCCIÓN**

* GOR: Relación gas/petróleo producido.
* Pet_1A: cantidad acumulada de petróleo durante el primer año de producción del pozo.
* Gas_1A: cantidad acumulada de gas durante el primer año de producción del pozo.	
* Pet_2A: cantidad acumulada de petróleo despues de dos años de producción del pozo.	
* Gas_2A: cantidad acumulada de gas despues de dos años de producción del pozo.

* TipoPozo: tipo de pozo (gasifero, petrolifero, sumidero, otro).	
* TipoExtraccion: método de extracción del fluido (surgencia natural, cavidad progresiva, gas lift, bombeo mecanico).		
* Diseno_Pozo: tipo de diseño del pozo (horizontal, vertical o dirigido)
